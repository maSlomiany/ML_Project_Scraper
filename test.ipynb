# Serper

```python
import http.client
import json

def serch_person(person: str) -> dict:

    conn = http.client.HTTPSConnection("google.serper.dev")
    payload = json.dumps({
    "q": person,
    "num": 100
    })
    headers = {
    'X-API-KEY': '21d558dd45bd0b69bc9f70bb3c89e2554c97fc6a',
    'Content-Type': 'application/json'
    }
    conn.request("POST", "/search", payload, headers)
    res = conn.getresponse()
    data = res.read()
    return json.loads(data.decode("utf-8"))
```

```python
results = serch_person(politicians[0])
print(type(results))
```

# Scraping

```python
search_results = list()
with open('search_output.jsonl', 'r') as file:
    for line in file:
        search_results.append(json.loads(line))
```

```python
from scraper import scrape_body_text
```

```python
count_good = 0
count_bad = 0
with open('scraper_results.jsonl','a') as file:
    for result in search_results[0]['results']['organic']:
        scraped_content = scrape_body_text(result['link'])
        output_dict = {'person': search_results[0]['person'], 'url': result['link'], 'contents': scraped_content}
        if scraped_content:
            file.write(json.dumps(output_dict) + '\n')
            count_good += 1
        else:
            count_bad += 1
print(f"Scarped: {count_good}. Not scraped: {count_bad} ")
```
